```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Auto reloads notebook when changes are made
%reload_ext autoreload
%autoreload 2
```

```python
sns.set(style = 'darkgrid')
```

```python
print(sns.get_dataset_names()) #This shows the built in dataset 
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset
```

```python
dataset.tail()
```

```python
sns.relplot(data =dataset, x="total_bill",y="tip",hue="sex") #Relation between x and y according to hue here sex
```

```python
sns.relplot(data =dataset, x="total_bill",y="tip",hue="sex",col ="day")
```

```python
sns.relplot(data =dataset, x="total_bill",y="tip",hue="sex",col ="day",row='time')
```

## <span style = "color:red"> How to play with line plots in seaborn </span>

```python
newdata = sns.load_dataset('flights')
```

```python
newdata
```

```python
newdata.head()
```

```python
newdata.tail()
```

```python
flightData = newdata.pivot("year","month","passengers")
flightData.tail()
```

```python
sns.lineplot(data = flightData)
```

```python
sns.lineplot(data = newdata,x = "year",y="passengers")
```

```python
sns.lineplot(data = flightData["May"])
```

```python
sns.lineplot(data = newdata,x="year",y="passengers",hue ="month")
```
