### <span style = "color:blaack"> This tutorial gives overall ideas of how to use seaborn for distribution plots such as line, scatters etc. </span>|

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Auto reloads notebook when changes are made
%reload_ext autoreload
%autoreload 2
```

```python
print(sns.get_dataset_names())
```

```python
dataset = sns.load_dataset("penguins")
```

```python
dataset.tail()
```

### <span style = "color:red"> Let's play with histogram plots > </span>

```python
sns.histplot(data = dataset,x="island").set_title("Penguins")
```

```python
sns.histplot(data =dataset,x = "flipper_length_mm",binwidth=10,kde=True).set_title("Penguines")
```

```python
sns.histplot(data =dataset,x = "flipper_length_mm",hue = "island",binwidth=10,kde=True).set_title("Penguines")
```

## <span style = "color: red">Lets' play with KDE kernal density estimation </span>

```python
tips = sns.load_dataset("tips")
```

```python
tips.tail()
```

```python
sns.kdeplot(data=tips,x="total_bill",hue = "time",kernel = "gaussian").set_title("KDE PLOT")
```

## <span style = "color:red"> Lets' play with displot </span>

```python
sns.displot(data = dataset,x="flipper_length_mm",hue = "island",kde = True,col = "species")
```

# <span style = "color:blue"> How to play with distplot in seaborn </span>

```python
from scipy.stats import norm
sns.distplot(tips['total_bill'],fit=norm,kde=True,kde_kws={"color":"red","label": "KDE"},axlabel="bill").set_title("distribution plot")
```
